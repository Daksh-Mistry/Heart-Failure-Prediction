```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart.csv')
df.head()
```

```python
df['RestingECG'].value_counts()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['ChestPainType_Encoded'] = encoder.fit_transform(df['ChestPainType'])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['ExerciseAngina_Encoded'] = encoder.fit_transform(df['ExerciseAngina'])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Sex_Encoded'] = encoder.fit_transform(df['Sex'])
```

```python
df.drop(columns=['ChestPainType','Sex','ExerciseAngina'], inplace = True)
```

```python
df.head()
```

```python
# df.drop( columns = ['Sex_Encoded','FastingBS','ST_Slope', 'Oldpeak'], inplace =True)
```

```python
df_final = pd.get_dummies(df, columns=['RestingECG','ST_Slope'], dtype=int)

df_final.head()
```

```python
from sklearn.model_selection import train_test_split
y = df_final['HeartDisease']
df_final.drop(columns=['HeartDisease'], inplace = True)
X = df_final
X_train, X_test ,y_train ,y_test = train_test_split(X,y, test_size = 0.20, random_state=20)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train[['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']] = scaler.fit_transform(X_train[['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']])
X_test[['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']] = scaler.transform(X_test[['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']])
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```
